# show the top-percent Complaint Type in a pie chart

```python
import matplotlib.pyplot as plt
def my_autopct(pct):
    if pct > 5:
        return ('%.2f' % pct)
    else:
        return ''
my_label = list(df['Complaint Type'].value_counts().index[:4])
my_label = my_label +[None]*181
plt.pie(df['Complaint Type'].value_counts(), autopct=my_autopct, labels= my_label)
plt.show()
```
