```python
import pandas as pd
import numpy as np
datafile = "311_Service_Requests_2020.csv"
df = pd.read_csv(datafile)
```

```python
df.columns
```

Analyze what the top 10 causes of calls to 311 are in 11101

```python
df[df['Incident Zip'] == 11101]['Complaint Type'].value_counts()[:10]
```

Show the top-percent Complaint Type of area 11101 in a pie chart

```python
import matplotlib.pyplot as plt
def my_autopct(pct):
    if pct > 5:
        return ('%.2f' % pct)
    else:
        return ''
my_label = list(df[df['Incident Zip'] == 11101]['Complaint Type'].value_counts().index[:5])
my_label = my_label +[None]*126
plt.pie(df[df['Incident Zip'] == 11101]['Complaint Type'].value_counts(), autopct=my_autopct, labels= my_label)
plt.show()
```
